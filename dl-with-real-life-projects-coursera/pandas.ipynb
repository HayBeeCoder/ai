# Merging or Concantenating two dataframes

```python
dataframe_1 = pd.DataFrame([['a', 1],['b',2]], columns=['col1', 'number'])
```

```python
dataframe_2 = pd.DataFrame([["c", 3, "lion"], ["d",4,"tiger"]], columns=["letter","number", "animal"])
```

```python
dataframe_1
```

```python
dataframe_2
```

```python
pd.concat([dataframe_1, dataframe_2], axis=0)
```

```python
pd.concat([dataframe_1, dataframe_2], axis=1)
```

```python
pd.concat([dataframe_1, dataframe_2], axis=0, ignore_index=True)
```

```python
dataframe_1 = pd.DataFrame({
    "city":  ["Ogun", "Lagos", "Ibadan", "Kwara"],
    "temperature": [23,24,25,28]
})
```

```python
dataframe_1
```

```python
dataframe_2 = pd.DataFrame({
    "city":  [ "Lagos", "Ibadan", "Kwara"],
    "temperature": [10,12,15]
})
```

```python
dataframe_2
```

```python
pd.merge(dataframe_1, dataframe_2, on="city")
```

```python
pd.merge(dataframe_1, dataframe_2, how="outer")
```

```python
pd.merge(dataframe_1, dataframe_2, how="left")
```

```python
pd.merge(dataframe_1, dataframe_2, how="right")
```

```python
dataframe_3 = pd.DataFrame({
    "city":  [ "Lagos", "Ibadan", "Kwara"],
    "humidity": [10,12,15]
})
```

```python
pd.merge(dataframe_1, dataframe_3, on="city")
```

```python
pd.merge(dataframe_1, dataframe_3, on="city", how="outer")
```

```python
pd.merge(dataframe_1, dataframe_3, on="city", how="left")
```

```python
pd.merge(dataframe_1, dataframe_3, on="city", how="right")
```
