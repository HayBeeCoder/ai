```python
sns.boxplot(y='Sales',x='Product_Category', data=df)
plt.yscale('log')
plt.show()
```

```python
sns.boxplot(y='Profit',x='Product_Category', data=df)
plt.yscale('log')
plt.show()
```

```python
df = df[(df.Profit > -1000) & (df.Profit < 1000)]

sns.boxplot(y='Profit',x='Product_Category', data=df)
# plt.yscale('log')
plt.show()
```
