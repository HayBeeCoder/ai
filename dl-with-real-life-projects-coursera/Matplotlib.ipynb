```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

x = np.linspace(5,100,100)
y = np.linspace(10, 1000, 100 )

plt.plot(x,y)
plt.show()
```

```python
plt.plot([1,3,5,6],[3,9,6,1])
plt.show()
```

```python
plt.plot(x,y)

plt.xlabel("Current")
plt.ylabel("Voltage")
plt.title("Ohm's law")

# plt.xlim([20,80])
# plt.ylim([200,800])

plt.axis((20,80,200, 800))
plt.show()
```

```python
x = np.linspace(0,10,20)
y = x**2

plt.plot(x,y,'b+')

plt.xlabel("Current")
plt.ylabel("Voltage")
plt.title("Ohm's law")
```

```python
x = np.linspace(0,20,10)
y = np.linspace(2,8, 10)


plt.plot(x,y,"b-", x, y ** 2, 'r^', x, y**3, 'g+')
plt.show()
```

```python
# figures and subplots
```

```python
x = np.linspace(1,20,10)
y = np.log(x)

plt.figure(1)
plt.subplot(121)
plt.xlabel("x")
plt.ylabel("log y")
plt.title("Graph of log y against x")
plt.plot(x, y)

plt.subplot(122)
plt.xlabel("x")
plt.ylabel("log(y) ** 2")
plt.title("Graph of log(y)^2 against x")
plt.plot(x, y ** 2)


plt.subplots_adjust(wspace=0.3)
```

```python
x = np.linspace(1,2,10)

plt.figure(1)
plt.subplot(121)
plt.title("Linear")
plt.plot(x, x)

plt.subplot(122)
plt.title("Cubic")
plt.plot(x, x ** 3)

plt.figure(2)
plt.subplot(221)
plt.title("Log")
plt.plot(x, np.log(x))

plt.subplot(222)
plt.title("Exponential")
plt.plot(x,np.exp(x))

plt.subplots_adjust(wspace=0.3)
```
